```python
'''
Assignment operator
==
+=
-=
*=
/=
//=
%=
**=

'''
```

```python
a = 10

a += 20  # a = a+20
print(a)
```

```python
b = 10
b -= 20  # a = a - 10
print(b)

b *= b
print(b)

a = 10
b /= a   # b = b/a == 100 / 10
print(b)

a //= a
print(a)

a %= a
print(a)

a = 10
a **= a
print(a)
```

```python
'''
Comparison operator(True / False)
==
!=
>
<
>=
<=
'''
```

```python
a = 10
b = 20

print(a == 10)
print(a <= b)
print(a >= 15)
print(a != b)
```

```python
'''Membership operator
In
Not In'''

a = ["Apple", "Orange", "Mango", "Kiwi"]
print(a)

print("Kiwi" in a)

a = "Hello"
print("H" in a)
print("Hl" in a)

a = "123"
print("2" in a)

print("apple" not in a)
print(a)
```

```python
'''
Logical operator
And
Or
Not
'''
```

```python
a = "Alex"
b = "Joe"

print(a == "Alex" and b == "joe")
print(a == "Alex" or b == "joe")
print(not(a == "alex"))
print(not(a != "Alex"))
```

```python
'''Identity operator
Is
Is not'''

a = ["Apple", "Orange", "Mango", "Kiwi"]
b = ["Apple", "Orange", "Mango", "Kiwi"]

print(a is b)

x = a
print(x)

print(x is a)

y = list(a)
print(y)

print(a is y)
```

```python
#Format String

name = "Alex"
office = "Aimore"
salary = 40000

txt = "my name is {}, im working in {} and my slary is {}"
print(txt.format(name, office, salary))

txt = "my name is {1}, im working in {0} and my slary is {2}, Arun also working in {0}"
print(txt.format(office, name, salary))
#                 0        1    2
```

```python
#f String
name = "Alex"
office = "Aimore"
salary = 40000

txt = f"my name is {name}, im working in {office} and my slary is {salary}, Arun also working in {office}"
print(txt)
```

```python
#Escape character:
'''
\'
\"
\\
\t
\b
\n
\r
'''
```

```python
txt = "my name is \'Alex\'"
print(txt)

txt = "my name is \"Alex"
print(txt)

txt = "my name is \\Alex"
print(txt)

txt = "my name is \tAlex"
print(txt)

txt = "my name is \bAlex"
print(txt)

txt = "my \nname \nis \nAlex"
print(txt)

txt = "my name is \rAlex"
print(txt)
```

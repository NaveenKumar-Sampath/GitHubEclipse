```python
class new:
    def add(self,a,b):
        c = a+b
        print(c)

    def sub(self,a,b):
        c = a-b
        print(c)

n = new()
n.add(20,20)
n.sub(10,20)
```

```python
class car:
    def __init__(self, color, brand):
        self.color = color
        self.brand = brand

    def drive(self):
        print(f"Im driving a {self.color} color, {self.brand} Car")
        
    def colors(self, a):
        print(f"my car colour was {self.color}")
        print(a)

c = car("Red","RR")
c.drive()
c.colors(20)

Naveen = car("Black","BMW")
Naveen.drive()
Naveen.colors(30)
```

```python
#Single Inheritance:
#--A child class inherit from one parent class.

class father:
    def bike(self):
        print("Father has Bike")

class son(father):
    def car(self):
        print("Son has a car")


s = son()
s.bike()
s.car()
```

```python
#Multiple:
#--A child class inherits from multiple parent class.

class father:
    def bike(self):
        print("Father has Bike")

class mother:
    def scooty(self):
        print("Mother has scotty")

class son(father, mother):
    def car(self):
        print("Son has a car")

s = son()
s.bike()
s.scooty()
s.car()
```

```python
#Multilevel:
#--Invlove a chain of inheritance where a child class become a parent


class gfather:
    def bulletcart(self):
        print("GFather has Bulletcart")

class father(gfather):
    def bike(self):
        print("Father has Bike")

class son(father):
    def car(self):
        print("Son has a car")

class Gson(son):
    def flight(self):
        print("Gson has flyt")

gs = Gson()
gs.bulletcart()
gs.bike()
gs.car()
gs.flight()
```

```python
#Hierarchical:
#--Hierarchical inheritance occurs when multiple child class inherts from a single parent class.


class father:
    def bike(self):
        print("Father has Bike")

class son(father):
    def car(self):
        print("Son has a car")

class daughter(son):
    def scooty(self):
        print("daughter has scooty")

s = son()
d = daughter()


d.bike()
d.scooty()
d.car()
```

```python
#Hybrid:
class GrandParent:
    def house(self):
        print("Grandparent has a house")

class Parent1(GrandParent):
    def car(self):
        print("Parent1 has a car")

class Parent2(GrandParent):
    def land(self):
        print("Parent2 has land")

class Child(Parent1, Parent2):
    def bike(self):
        print("Child has a bike")
```

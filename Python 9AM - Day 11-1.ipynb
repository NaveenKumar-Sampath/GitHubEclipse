```python
a = 0

while a < 10:
    a += 1
    print(a)
```

```python
#Break:
#Else:
#When the condition longer is True

i = 0
while i < 6:
    
    if i == 3:
        break
    i += 1
    print(i)
else:
    print("While is finsihed")
```

```python
i = 0

while i < 6:
    i += 1
    if i == 3:
        continue  
    print(i)
else:
    print("While is finsihed")
```

```python
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        super().sound()
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")
        
d = Animal()
d.sound()
        
"""Polymorphic behavior
for animal in [Dog(), Cat()]:
    animal.sound()"""
```

```python
class Calculator:
    def add(self, a):
        return a 

    def add(self, a, b):
        return a + b 

    def add(self, a, b, c):
        return a + b + c

c = Calculator()
c.add(20, 20, 30)
```

```python
#Encapsulation
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private variable

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ₹{amount}, New balance: ₹{self.__balance}")
        else:
            print("Insufficient balance")

    def check_balance(self):
        return self.__balance

ba = BankAccount(1000)
ba.withdraw(5000)
ba.check_balance()
```

```python
from abc import ABC, abstractmethod  	#Abstract Base Class

class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        print("Bark")

class Cat(Animal):
    def make_sound(self):
        print("Meow")

d = Dog()
d.make_sound()  # Bark

c = Cat()
c.make_sound()  # Meow
```

```python
import csv

with open("C:\\Users\\aimor\\Desktop\\students.csv", "r") as f:
    reader = csv.reader(f)
    next(reader)  #skip header
    for row in reader:
        name = row[0]
        grade = int(row[1])

        if grade >= 50:
            print(f"{name} Passed")
        else:
            print(f"{name} Failed")
```

```python
total = 0
with open("C:\\Users\\aimor\\Desktop\\product.csv", "r") as f:
    reader = csv.reader(f)
    next(reader)  #skip header
    for row in reader:
        total += float(row[1])

print("Total sales :", total)
```

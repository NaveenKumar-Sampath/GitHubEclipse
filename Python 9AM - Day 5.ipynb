```python
#List: []

a = ["apple", "mango","orange", 1, True, False, 0, 23, 56.9,"apple"]
print(a)
print(type(a))
```

```python
print("Apple" in a)
```

```python
#Change the item:
a = ["apple", "mango","orange", "kiwi"]
a[1] = "Guva"
print(a)
a[0:3] = "Pineappne"
print(a)

'''
P i n e a p p n e
0 1 2 3 4 5 6 7 8'''
```

```python
a = ["apple", "mango","orange", "kiwi"]
a[0:3] = ["Pineappne","Graps"]
print(a)
```

```python
#Insert new item:
a = ["apple", "mango","orange", "kiwi"]
a.insert(0,"Graps")
print(a)
```

```python
a.insert(7,"Graps")
print(a)
```

```python
#append:  --Add item end of the list
a = ["apple", "mango","orange", "kiwi"]
a.append("Graps")
print(a)
b = [1,2,3,4,5]
a.append(b)
print(a)
```

```python
#Extend: 
a = ["apple", "mango","orange", "kiwi"]
b = [1,2,3,4,5]
a.extend(b)
print(a)
```

```python
#Remove / pop:

a = ["apple", "mango","orange", "kiwi", "apple","Apple"]
a.remove("apple")
print(a)
```

```python
a.pop(1)
print(a)
a.pop()
print(a)

a = ["apple", "mango","orange", "kiwi", "apple","Apple"]
a.pop([0:3])
print(a)
```

```python
#Clear:
a = ["apple", "mango","orange", "kiwi", "apple","Apple"]
a.clear()
print(a)
print(type(a))
```

```python
#delete:
a = ["apple", "mango","orange", "kiwi", "apple","Apple"]
del a
print(a)
```

```python
#Sorting: asc / desc:
a = ["apple", "mango","Orange", "kiwi","Kiwi" ,"apple","Apple","Zoo","Xyz"]
a.sort()
print(a)
b = [100, 47, 97, 12, 90, 10, 8, 3, 1, 0]
b.sort()
print(b)
```

```python
a.sort(reverse = True)
print(a)
b.sort(reverse = True)
print(b)
```
